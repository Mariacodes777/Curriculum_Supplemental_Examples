### Below I will go over some examples similar to what is asked for your assignment

```r
# I want to look at all the countries in the country column:
unique(gapminder$country)
```

```r
#Now I can filter to select 5 countries:
Countries5 <-gapminder %>% filter(country %in% c("Madagascar", "Tanzania", "Peru", "Albania", "Pakistan"))
```

```r
#In the above code, you can see that we are starting out specifying the dataset name, which is gapminder, then we
#use the 'and then' %>% and the filter() function stating the column we want to filter inside the (), so we put
#country since we want to filter that column to select our 5 countries. Next, we use the %in% and place a lower case
#c with () to start our vector of countries, and select whichever 5 countries we like with them inside "" separated
#by commas :) Notice that we have two sets of(), one for the filter() and one for the vector c()
```

### Let's explore which of the 5 countries has the lowest population in 1982 and 2007:

```r
#Below, I named my filtered data LowPoP, I used Countries5 which is our selected 5 countries. I did year in (%in%) 
#starting a vector for the years I wanted to look at, 1982 and 2007, then I did 'and then' sign: %>% to arrange
#my data by the year and selected which columns I wanted to look at using the select() function and putting the 
#3 columns of interest to me inside the ()
```

```r
LowPoP <- filter(Countries5, year %in% c(1982, 2007)) %>% arrange(year) %>% select(country, pop, year)
```

```r
LowPoP
```

```r
#Above we can see the populations by year for the 5 countries and can see that in 1982 & 2007 Albania had the 
#lowest population and Pakiston had the highest.
```

### Let's do a line plot with per cap and life expectancy and have a different color represent each country

```r
#In the code below, I am using the ggplot() function to start this plot, then I specify the dataset name, which
#for us is Countries5 since we only want to plot this for our 5 selected countries. On the x axis, the gdpPercap
#is plotted and on the y axis the lifeExp is plotted and the color= country is saying that we want each country's
#line plot to be in a different color. geom_line() is used to indicate a line plot.
```

```r
ggplot(Countries5, aes(x= gdpPercap, y= lifeExp , color= country)) + geom_line()
```

```r
#There would need to be more research as to why, but from the graph it does look like higher gdpPercap is 
#showing higher life expectancy. This sadly makes sense as less funds can lead to less resources, medical care, 
#etc. We can see the interesting line for Albania and Pakistan, while Tanzania, had the least gdpPercap and the 
#lowest life expectancy.
```

### Now we will look at the average gdpPercap for the 5 countries and how that compares to the average of the other countries in the gapminder dataset.

```r
#This code below will be similar to a question asked on your assigment where you are asked about the median. 
#You see how we will start by giving it a name, I am calling this Average5C because we are looking for the average
#gdpPercap for the 5 Countries. Then we will specify the %>% and select() which columns we want to see. Here we are 
#going to look at gdpPercap and country. Then we will use %>% and group_by() country, and %>% summarise()
#here is where we will specify a name for what we are about to do, I said Average= since I want the mean.
#Then I state the function I want computed, in this case it is mean(gdpPercap) column.
```

```r
Average5C <- Countries5 %>% select( gdpPercap, country) %>% group_by(country) %>% summarise(Average = mean(gdpPercap))
```

```r
Average5C
```

```r
#Tanzania had the lowest and Peru had the highest
```

```r
#Now we will look at the average of the whole gapminder dataset and see how it compares
```

```r
AverageGapminder <- gapminder %>% select( gdpPercap, country) %>% group_by(country) %>% summarise(Average = mean(gdpPercap)) %>% arrange(desc(Average))
```

```r
AverageGapminder
```
